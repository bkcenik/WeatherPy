# Weatherpy

### 1. Import dependencies

```python
# Dependencies
import json
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from config import api_key
from pprint import pprint
from citipy import citipy
import time
```

### 2.Randomly sample cities.

```python
# random sampling of geocoordinates
# ranges
lats_ran = (-90, 90)
lngs_ran = (-180, 180)

# make empty lists
lat_lng = []
cities = []

# sample random latitude and longitudes, zip them together
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lng = zip(lats, lngs)
```

```python
# create list of cities
for coordinate in lat_lng:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name
    if city not in cities:
        cities.append(city)
```

### 3. Create dataframe, query OpenWeather API to fill in information.

```python
# create empty dataframe
weatherpy_df = pd.DataFrame({"City": cities})
weatherpy_df["Cloudiness"] = ""
weatherpy_df["Country"] = ""
weatherpy_df["Date"] = ""
weatherpy_df["Humidity"] = ""
weatherpy_df["Lat"] = ""
weatherpy_df["Lng"] = ""
weatherpy_df["Max Temp"] = ""
weatherpy_df["Wind Speed"] = ""

weatherpy_df.head()
```

```python
# perform API calls and iterate through rows and indices to fill dataframe
#base url
url = f"http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID={api_key}&q="

for index, row in weatherpy_df.iterrows():
    city = row["City"]
    response = requests.get(url + city).json()
    
    try:
        print(f"Searching {index + 1} of {len(cities)} items | Name of city: {city}")
        print(url + city)
        
        weatherpy_df.loc[index, "Cloudiness"] = response["clouds"]["all"]
        weatherpy_df.loc[index, "Country"] = response["sys"]["country"]
        weatherpy_df.loc[index, "Date"] = response["dt"]
        weatherpy_df.loc[index, "Humidity"] = response["main"]["humidity"]
        weatherpy_df.loc[index, "Lat"] = response["coord"]["lat"]
        weatherpy_df.loc[index, "Lng"] = response["coord"]["lon"]
        weatherpy_df.loc[index, "Max Temp"] = response["main"]["temp_max"]
        weatherpy_df.loc[index, "Wind Speed"] = response["wind"]["speed"]
        
    except (KeyError, IndexError):
        print("Couldn't find city. Next!")
```

```python
weatherpy_df.head(30)
```

```python
# some cities do not have countries (these are the ones that are "not found")
#Drop cities with no matching countries
weatherpy_df_cleaned = weatherpy_df[weatherpy_df["Country"] != ""] # set to "" since empty countries are not designated as NaN

weatherpy_df_cleaned.count() # to make sure there are at least 500 cities
```

```python
weatherpy_df_cleaned.head(30)
```

```python
# export as csv file
weatherpy_df_cleaned.to_csv("weatherpy_data.csv")
```

### 4. Plot data.

* Temperature versus Latitude

```python
plt.scatter(weatherpy_df_cleaned["Lat"], weatherpy_df_cleaned["Max Temp"], marker="o", facecolor="pink", edgecolor="black")
date = time.strftime("%m/%d/%Y")
plt.title(f"City Latitude vs Temperature ({date})")
plt.ylabel("Latitude")
plt.xlabel("Temperature (F)")
plt.grid(True)
plt.savefig("TempLat.png")
```
